```python
dataset_path = "/kaggle/input/fall-detection-le2i/LE2I-subset-yolov11"
train_dir = f"{dataset_path}/train/images"
val_dir = f"{dataset_path}/valid/images"
test_dir = f"{dataset_path}/test/images"
data_yaml = f"{dataset_path}/data.yaml"
```

```python
from ultralytics import YOLO
```

```python
model = YOLO('yolo11n.pt')#using nano
```

```python
results = model.train(
    data=data_yaml,
    epochs=100,
    imgsz=640,
    batch=16,
    project = 'fall_detection',
    name = 'yolov11 model',
    verbose = True,
    save = True,
    patience = 0,
    optimizer = 'Adam',
    lr0 = 0.001)
```

```python
# Validate the model
metrics = model.val()
print(metrics)
```

```python
best_model = YOLO('/kaggle/working/runs/detect/train/weights/best.pt')
```

```python
import os,cv2

test_folder = '/kaggle/input/fall-detection-le2i/LE2I-subset-yolov11/test/images'  # Replace with your test folder path
output_folder = '/kaggle/working/all_epochs_test_annotated_frames'  # Folder to save the annotated frames

# Create output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Iterate through all the test images in the folder
for filename in os.listdir(test_folder):
    if filename.endswith(".jpg") or filename.endswith(".png"):  # Adjust based on your image formats
        image_path = os.path.join(test_folder, filename)

        # Read the image
        image = cv2.imread(image_path)

        # Predict on the image
        results = best_model(image)

        # Annotate the image with predictions
        annotated_image = results[0].plot()  # Annotate the frame

        # Save the annotated image to the output folder
        output_image_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_image_path, annotated_image)

        print(f"Processed {filename}")

print("Testing completed on all frames.")
```

```python


# Input and output folder paths
input_folder = '/kaggle/input/fall-detection-test-videos/test videos'  # Replace with your uploaded folder path

output_folder = '/kaggle/working/annotated_videos/'
os.makedirs(output_folder, exist_ok=True)

# Process each video in the input folder
for video_name in os.listdir(input_folder):
    if video_name.endswith(('.mp4', '.avi', '.mov', '.mkv')):  # Check for video formats
        video_path = os.path.join(input_folder, video_name)
        print(f"Processing video: {video_name}")

        # Open the video file
        cap = cv2.VideoCapture(video_path)

        # Get video properties
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for MP4
        fps = int(cap.get(cv2.CAP_PROP_FPS))
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

        # Output video path
        output_video_path = os.path.join(output_folder, f"annotated_{video_name}")
        out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

        # Process each frame in the video
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            # Run YOLO model prediction on the frame
            results = best_model(frame)

            # Annotate the frame with predictions
            annotated_frame = results[0].plot()

            # Write the annotated frame to the output video
            out.write(annotated_frame)

        # Release resources for this video
        cap.release()
        out.release()

        print(f"Annotated video saved: {output_video_path}")

print("Processing complete. Annotated videos are saved in:", output_folder)
```
